```python
import pandas as pd
import openpyxl
import numpy as np
import re
```

```python
file_path = 'Mid.xlsx'
mid = pd.read_excel("Raw Data\Mid.xlsx")

def remove_non_thai(text):
    # ใช้ regex เพื่อหาเฉพาะข้อความภาษาไทยและวรรคตอน
    thai_pattern = re.compile(r'[ก-๙\s]+')
    thai_text = ' '.join(thai_pattern.findall(str(text)))
    return thai_text

# นำฟังก์ชันไปใช้กับคอลัมน์ที่ต้องการ
mid['Unnamed: 0'] = mid['Unnamed: 0'].apply(remove_non_thai)

# บันทึกไฟล์ใหม่
output_file_path = 'Mid_Cleaned.xlsx'
mid.to_excel(output_file_path, index=False)

print(f"ไฟล์ที่ทำความสะอาดแล้วถูกบันทึกไว้ที่: {output_file_path}")
mid
```

```python
south = pd.read_excel("Raw Data\South.xlsx")
north = pd.read_excel("Raw Data/North.xlsx")
esan = pd.read_excel("Raw Data\ESAN.xlsx")
```

```python
esan.rename(columns={"Unnamed: 0": "text"}, inplace=True)
north.rename(columns={"Unnamed: 0": "text"}, inplace=True)
mid.rename(columns={"Unnamed: 0": "text"}, inplace=True)
south.rename(columns={"Unnamed: 0": "text"}, inplace=True)
mid
```

```python
south = south.drop_duplicates(keep='first')
south['text'] = south['text'].apply(remove_non_thai)

north = north.drop_duplicates(keep='first')
north['text'] = north['text'].apply(remove_non_thai)
north

mid = mid.drop_duplicates(keep='first')


esan = esan.drop_duplicates(keep='first')
esan['text'] = esan['text'].apply(remove_non_thai)
```

```python
import pandas as pd
from pythainlp import word_tokenize
from nltk.util import ngrams

def ng(x):
    all_ngrams = []
    text_column = x['text'].astype(str)
    for text in text_column:

            words = word_tokenize(text, engine='newmm')
            for n in range(1, 4):
                n_grams = list(ngrams(words, n))
                all_ngrams.extend([' '.join(gram) for gram in n_grams])
            all_ngrams.append(text)
    return all_ngrams
```

```python
sd = ng(south)
south_df = pd.DataFrame(sd, columns=['text'])
south_df = south_df.drop_duplicates(keep='first')
south_df.reset_index(drop=True, inplace=True)
south_df
```

```python
md = ng(mid)
mid_df = pd.DataFrame(md, columns=['text'])
mid_df = mid_df.drop_duplicates(keep='first')
mid_df.reset_index(drop=True, inplace=True)
mid_df
```

```python
nd = ng(north)
north_df = pd.DataFrame(nd, columns=['text'])
north_df = north_df.drop_duplicates(keep='first')
north_df.reset_index(drop=True, inplace=True)
north_df
```

```python
es = ng(esan)
esan_df = pd.DataFrame(es, columns=['text'])
esan_df = esan_df.drop_duplicates(keep='first')
esan_df.reset_index(drop=True, inplace=True)
esan_df
```

```python
south_df = pd.DataFrame(south_df,columns=['text'])
south_df['label'] = "s"
north_df = pd.DataFrame(north_df,columns=['text'])
north_df['label'] = "n"
mid_df = pd.DataFrame(mid_df,columns=['text'])
mid_df['label'] = "m"
esan_df = pd.DataFrame(esan,columns=['text'])
esan_df['label'] = "e"
```

```python
df = pd.concat([south_df, north_df, mid_df,esan_df])
np.random.seed(42)
df.reset_index(drop=True, inplace=True)
```
